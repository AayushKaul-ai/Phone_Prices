```python
import pandas as pd
```

```python
df=pd.read_csv('mobile_train.csv')
```

```python
df.info()
```

```python
#Separate X and Y
x=df.drop(labels='price_range',axis=1)
y=df[['price_range']]
```

```python
x
```

```python
y
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
# Train test split
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Building Networks
from keras.models import Sequential
from keras.layers import Input,Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
xtrain.shape
```

```python
from keras.utils import to_categorical
```

```python
# For Multiclass
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
#Initialize neural network
model=Sequential()
```

```python
ytrain.shape
```

```python
# Input Model
model.add(Input(shape=(xtrain.shape[1],)))
```

```python
# Add Hidden Layer
model.add(Dense(units=64,activation='relu'))
# Add dropout layer
model.add(Dropout(rate=0.3))
```

```python
# Add Hidden layer 2
model.add(Dense(units=32,activation='relu'))
# Add dropout layer
model.add(Dropout(rate=0.2))
```

```python
# Add Output Layer
model.add(Dense(units=ytrain_cat.shape[1],activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
es=EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[es])
```

```python
# Curve
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Trainingloss')
plt.plot(hist.history['val_loss'],label='Validationloss')
plt.xlabel('Epochs')
plt.ylabel('Categorical Crossentropy')
plt.title('Learning Curve')
plt.show()
```

```python
model.evaluate(xtrain,ytrain_cat)
```

```python
tr_prob=model.predict(xtrain)
ts_prob=model.predict(xtest)
```

```python
model.evaluate(xtest,ytest_cat)
```
